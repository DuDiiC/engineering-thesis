```python
import pandas as pd
```

```python
watchers = pd.read_pickle('../../data/01_data_from_db/watchers.pkl')
watchers['year'] = watchers['created_at'].dt.year
watchers['month'] = watchers['created_at'].dt.month
watchers.head()
```

Liczba nowych obserwujących w miesiącu dla każdego projektu

```python
new_watchers = watchers.groupby(['project_id', 'year', 'month']).count().reset_index() \
    .rename(columns={'created_at': 'new_watchers'})
new_watchers.head()
```

Łączna liczba obserwujących dla każdego projektu na początku danego miesiąca

```python
total_watchers = []
for index, row in new_watchers.iterrows():
    total_watchers.append(new_watchers[(row['project_id'] == new_watchers['project_id']) & 
            ((row['year'] > new_watchers['year']) | 
             ((row['year'] == new_watchers['year']) & 
              (row['month'] > new_watchers['month'])))]['new_watchers'].sum())
```

```python
new_watchers['total_watchers'] = pd.Series(total_watchers)
```
