```python
import pandas as pd
```

```python
pull_requests = pd.read_pickle('../../data/01_data_from_db/pull_requests.pkl')
pull_requests.head()
```

```python
pull_requests_with_history = pd.read_pickle('../../data/01_data_from_db/pull_requests_with_history.pkl')
pull_requests_with_history.head()
```

```python
pull_requests_with_history['year'] = pull_requests_with_history['created_at'].dt.year
pull_requests_with_history['month'] = pull_requests_with_history['created_at'].dt.month
pull_requests_with_history
```

```python
grouped_pull_requests = pull_requests_with_history.groupby(['project_id', 'year', 'month', 'action', 'merged']) \
    .count().reset_index().dropna() \
    .rename(columns={'pull_request_id': 'new_pull_requests'}) \
    .drop(columns={'pull_request_history_id', 'created_at'})
grouped_pull_requests['new_pull_requests'] = grouped_pull_requests['new_pull_requests'].astype('int64')
grouped_pull_requests
```

```python
new_opened_pull_requests_to_merge = grouped_pull_requests[(grouped_pull_requests['action'] == 'opened') & (grouped_pull_requests['merged'] == 1)] \
    .rename(columns={'new_pull_requests': 'new_opened_pull_requests_to_merge'}) \
    .drop(columns={'action', 'merged'})
new_opened_pull_requests_to_merge
```

- liczba wszystkich zmergowanych pull requestów od początku projektu

// TODO: **CZY TUTAJ MA BYĆ AKCJA `MERGED` CZY `CLOSED`?**

```python
grouped_pull_requests_merged = grouped_pull_requests[(grouped_pull_requests['merged'] == 1) &
                                (grouped_pull_requests['action'] == 'merged')].drop(columns={'merged', 'action'})
grouped_pull_requests_merged
```

```python
total_merged_pull_requests = []
for index, row in grouped_pull_requests_merged.iterrows():
    total_merged_pull_requests.append(grouped_pull_requests_merged[
                    (row['project_id'] == grouped_pull_requests_merged['project_id']) & 
                    ((row['year'] > grouped_pull_requests_merged['year']) | 
                    ((row['year'] == grouped_pull_requests_merged['year']) &  
                    (row['month'] >= grouped_pull_requests_merged['month'])))]['new_pull_requests'].sum())
```

```python
grouped_pull_requests_merged['total_merged_pull_requests'] = total_merged_pull_requests
```
